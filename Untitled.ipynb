```python
%load_ext autoreload
%autoreload 2
from mesh import Tiling
from plates import Plates
import numpy as np
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
from numba import njit
import networkx as nx
from tqdm import tqdm
from dataclasses import dataclass
import itertools
from scipy.interpolate import interp2d
from light import Illumination
from scipy.integrate import solve_ivp
```

```python
class Atmosphere:
    def __init__(self,positions, sources, swirls,N):
        self.positions=np.array(positions)
        self.sources=np.array(sources)
        self.swirls=np.array(swirls)
        self.a=2/N
    def velocityField(self,positions):
        velocities=np.zeros_like(positions)
        for P_i,m_i,s_i in zip(self.positions,self.sources,self.swirls):
            P_i_long=np.tile(P_i,(len(positions),1))
            d=np.sum(P_i_long*positions,axis=-1)
            s=-m_i*(2*self.a)**.5*(P_i_long-d[:,None]*positions)/(1-d+self.a)[:,None]
            w=-s_i*(2*self.a)**.5*np.cross(P_i_long,positions)/(1-d+self.a)[:,None]
            velocities+=s+w          
        return velocities
    def update(self,duration):
        def f(t,positions):
            return self.velocityField(np.reshape(positions,np.shape(self.positions))).flatten()/planet.parameters.radius
        sol=solve_ivp(f,(0,duration),self.positions.flatten(),atol=1e-3,rtol=1e-4,t_eval=np.arange(0,duration,0.1))
        self.positions=np.reshape(sol.y[:,-1],np.shape(self.positions))
        self.positions/=np.linalg.norm(self.positions,axis=-1)[:,None]
    @property
    def numberOfSource(self):
        return len(self.positions)
            
```

```python
@dataclass
class PlanetParameters:
    radius: float
    starTemperature: float
    starRadius: float
    starDistance: float
    

class Planet:
    def __init__(self,parameters,seed=0):
        self.seed=seed
        self.parameters=parameters
    def generateTiling(self,numberTiles,iterateRelaxation=30):
        self.centers,self.corners=Tiling(self.seed).createGraph(numberTiles,iterate=iterateRelaxation)
    def generatePlates(self,numberPlates):
        self.p=Plates(self.centers)
        self.centers,self.plates=self.p.generatePlates(numberPlates)   
        self.centers,self.plates=self.p.randomSpeed(self.plates,self.centers) 
        self.centers,self.plates=self.p.platesElevation(self.plates,self.centers,flood=0.8)
        for plate in tqdm(self.plates,desc="Computing distance matrix"):
            for (source,targets) in nx.algorithms.shortest_paths.generic.shortest_path_length(plate,weight="length"):
                plate.nodes[source]["distances"]=targets
    def computeHeight(self):
        self.centers,self.plates=self.p.computePressure(self.plates,self.centers,self.corners,self.parameters)
        self.centers,self.plates=self.p.computeElevation(self.plates,self.centers,self.corners,self.parameters)
    def computeAnnualIllumination(self):
        tqdm.write("Computing annual illumination")
        lat,long=Projection().toLatLong(np.array([p for _,p in planet.centers.nodes.data("center")]),"deg")
        time=np.arange(0,365,0.1)
        light=Illumination(year=365,inclination=24/180*np.pi).light(long/180*np.pi,lat/180*np.pi,time)
        light=np.trapz(light,time,axis=-1)/(time[-1]-time[0])
        for i,l in enumerate(light):
            planet.centers.nodes[i]["illumination"]=l
    @property
    def numberPlates(self):
        return len(self.plates)
    @property
    def numberTiles(self):
        return self.centers.number_of_nodes()
    def getNearestTile(self,positions):
        M=np.zeros((len(self.lowerAtmosphere.positions),self.centers.number_of_nodes()))
        for ((i,p),(j,q)) in itertools.product(enumerate(positions),self.centers.nodes.data("center")):
            M[i,j]=np.arccos(np.dot(p,q))
        return np.argmin(M,axis=-1)
    def plotTiles(self,ax,contour=True,minLight=0.1):
        
        cmap_terrain = plt.cm.get_cmap('gist_earth')
        cmap_ocean = plt.cm.get_cmap('ocean')
        for i in tqdm(range(planet.centers.number_of_nodes()),desc="Plotting tiles"):    
            l=self.centers.nodes[i]["corners"]
            vertices=np.append(l,l[0])
            points=[]
            for j in vertices:
                points.append(self.corners.nodes[j]["position"])
            lat,long=Projection().toLatLong(np.array(points),"deg")
            if(self.centers.nodes[i]["elevation"]>0):
                y=np.interp(self.centers.nodes[i]["elevation"],[0,10],[0.5,1])
                color=cmap_terrain(y)
            else:
                y=np.interp(self.centers.nodes[i]["elevation"],[-10,0],[0.25,0.8])
                color=cmap_ocean(y)
            (r,g,b,a)=color
            illumination=min(1,self.centers.nodes[i]["illumination"]+minLight)
            color=(r*illumination,g*illumination,b*illumination,1)
            ax.fill(long,lat,color=color,transform=ccrs.Geodetic()) 
            if(contour):
                ax.plot(long,lat,color="black",transform=ccrs.Geodetic(),lw=1)
    def plotTemperatureMap(self,ax,contour=True,alpha=1,which="temperature"):
        heatMap = plt.cm.get_cmap('bwr')
        for i in tqdm(range(planet.centers.number_of_nodes()),desc="Plotting temperature map"):    
            l=self.centers.nodes[i]["corners"]
            vertices=np.append(l,l[0])
            points=[]
            for j in vertices:
                points.append(self.corners.nodes[j]["position"])
            lat,long=Projection().toLatLong(np.array(points),"deg")
            y=np.interp(self.centers.nodes[i][which]-273.15,[-15,40],[0,1])
            color=heatMap(y)
            ax.fill(long,lat,color=color,transform=ccrs.Geodetic(),alpha=alpha) 
            if(contour):
                ax.plot(long,lat,color="black",transform=ccrs.Geodetic(),lw=1)
    def plotLowerAirSpeed(self,ax):
        centers=np.array([c for _,c in self.centers.nodes.data("center")])
        lat,long=Projection().toLatLong(centers,"deg")
        velocities=planet.lowerAtmosphere.velocityField(centers)
        slat,slong=Projection().toLatLong(centers+velocities*100/self.parameters.radius,"deg")
        for i in tqdm(range(len(long)),desc="Plotting wind direction"):
            ax.plot([long[i],slong[i]],[lat[i],slat[i]],color="gray",transform=ccrs.Geodetic(),lw=1)   
    def plotHigherAirSpeed(self,ax):
        centers=np.array([c for _,c in self.centers.nodes.data("center")])
        lat,long=Projection().toLatLong(centers,"deg")
        upperAtmosphere=Atmosphere(self.lowerAtmosphere.positions,-self.lowerAtmosphere.sources,self.lowerAtmosphere.swirls,self.centers.number_of_nodes())
        velocities=upperAtmosphere.velocityField(centers)
        slat,slong=Projection().toLatLong(centers+velocities*100/self.parameters.radius,"deg")
        for i in tqdm(range(len(long)),desc="Plotting wind direction"):
            ax.plot([long[i],slong[i]],[lat[i],slat[i]],color="gray",transform=ccrs.Geodetic(),lw=1)  
    def plotPlatesSpeed(self,ax):
        centers=np.array([c for _,c in self.centers.nodes.data("center")])
        lat,long=Projection().toLatLong(centers,"deg")
        centerSpeeds=np.array([c for _,c in self.centers.nodes.data("speed")])
        slat,slong=Projection().toLatLong(centers+centerSpeeds/1000,"deg")
        for i in tqdm(range(len(long)),desc="Plotting plate velocities"):
            ax.plot([long[i],slong[i]],[lat[i],slat[i]],color="yellow",transform=ccrs.Geodetic(),lw=1)    
        ax.scatter(long,lat,color="orange",transform=ccrs.PlateCarree(),s=5,zorder=10)
    def plotPlates(self,ax):
        for (i,j) in tqdm(self.corners.edges,desc="Plotting plate borders"):
            (n1,n2)=self.corners.edges[i,j]["separates"]
            if self.centers.nodes[n1]["plate"]!=self.centers.nodes[n2]["plate"]:
                points=np.array([self.corners.nodes[i]["position"],self.corners.nodes[j]["position"]])
                lat,long=Projection().toLatLong(np.array(points),"deg")
                ax.plot(long,lat,color="red",transform=ccrs.Geodetic(),lw=2)
    def plotPressure(self,ax):
        for i,p in tqdm(self.centers.nodes.data("pressure"),desc="Plotting pressure"):
            if p is not None:
                lat,long=Projection().toLatLong(np.array([self.centers.nodes[i]["center"]]),"deg")
                if(p>0):
                    ax.scatter(long,lat,s=p/5,color="red",zorder=15,transform=ccrs.PlateCarree())
                else:
                    ax.scatter(long,lat,s=-p/5,color="blue",zorder=15,transform=ccrs.PlateCarree())
    def plotElevation(self,ax):
        points=np.array([p for _,p in self.centers.nodes.data("center")])
        elevation=np.array([e for _,e in self.centers.nodes.data("elevations")])
        lat,long=Projection().toLatLong(points,"deg")
        elevation_interp=interp2d(long,lat,elevation)
        long2d,lat2d=np.meshgrid(np.linspace(-180,180),np.linspace(-90,90))
        ax.contour(long2d,lat2d,elevation_interpolate(long2d,lat2d))
    def plotTemperature(self,ax):
        heatMap = plt.cm.get_cmap('bwr')
        lat,long=Projection().toLatLong(np.array([x for _,x in self.centers.nodes.data("center")]),"deg")
        for i in tqdm(range(len(lat)),desc="Plotting temperature"):
            y=np.interp(self.centers.nodes[i]["temperature"]-273.15,[-15,40],[0,1])
            color=heatMap(y)
            ax.text(long[i],lat[i],"{}".format(int(self.centers.nodes[i]["temperature"]-273.15)),transform=ccrs.PlateCarree(),color="lime",bbox=dict(boxstyle="circle",fc=color),ha="center",va="center",fontsize=5)
    def plotProperty(self,ax,prop):
        lat,long=Projection().toLatLong(np.array([x for _,x in self.centers.nodes.data("center")]),"deg")
        for i in tqdm(range(len(lat)),desc="Plotting {}".format(prop)):
            ax.text(long[i],lat[i],"{}".format(int(self.centers.nodes[i][prop])),transform=ccrs.PlateCarree(),color="lime",bbox=dict(boxstyle="circle",fc="orange"),ha="center",va="center",fontsize=5) 
```

```python
@njit
def toLatLong(r,ex,ey,ez):
    lat=np.zeros(r.shape[0])
    long=np.zeros(r.shape[0])
    for i in range(r.shape[0]):
        lat[i]=np.arccos(r[i][2]/np.sqrt(r[i][0]**2+r[i][1]**2+r[i][2]**2))
        long[i]=np.arctan2(r[i][1],r[i][0])
    return lat,long
def lengthSpherical(a,b):
    return np.arccos(np.sum(a*b,axis=-1))
def borderRoll(border):
    n=len(border)
    for i in range(n):
        yield((border[i],border[(i+1)%n]))
```

```python
from scipy.interpolate import griddata
centers=np.array([c for _,c in planet.centers.nodes.data("center")])

lat,long=Projection().toLatLong(centers,"deg")

dt=100
planet.lowerAtmosphere.update(dt)

for k,source in zip(planet.getNearestTile(planet.lowerAtmosphere.positions),planet.lowerAtmosphere.sources):
    n=planet.centers.nodes[k]
    if source<0:
        t=np.exp(source/2)
        n["upperAirTemperature"]=n["upperAirTemperature"]*t+(1-t)*n["lowerAirTemperature"]
    else:
        t=np.exp(-source/2)
        n["lowerAirTemperature"]=n["lowerAirTemperature"]*t+(1-t)*n["upperAirTemperature"]

lowerTemperatures=np.array([T for _,T in planet.centers.nodes.data("lowerAirTemperature")])
higherTemperatures=np.array([T for _,T in planet.centers.nodes.data("upperAirTemperature")])

velocities=planet.lowerAtmosphere.velocityField(centers)
slat,slong=Projection().toLatLong(centers-velocities*dt/planet.parameters.radius,"deg")

newTemp=griddata(np.array([[lat[i],long[i]] for i in range(len(lat))]),lowerTemperatures,np.array([[slat[i],slong[i]] for i in range(len(lat))]),method="cubic")
for i in planet.centers.nodes:
    if(not np.isnan(newTemp[i])):
        planet.centers.nodes[i]["lowerAirTemperature"]=newTemp[i]
        
higherAtmosphere=Atmosphere(planet.lowerAtmosphere.positions,-planet.lowerAtmosphere.sources,planet.lowerAtmosphere.swirls,planet.centers.number_of_nodes())
velocities=higherAtmosphere.velocityField(centers)
slat,slong=Projection().toLatLong(centers-velocities*dt/planet.parameters.radius,"deg")

newTemp=griddata(np.array([[lat[i],long[i]] for i in range(len(lat))]),higherTemperatures,np.array([[slat[i],slong[i]] for i in range(len(lat))]),method="cubic")
for i in planet.centers.nodes:
    if(not np.isnan(newTemp[i])):
        planet.centers.nodes[i]["upperAirTemperature"]=newTemp[i]

for k,(i,source) in zip(planet.getNearestTile(planet.lowerAtmosphere.positions),enumerate(planet.lowerAtmosphere.sources)):
    n=planet.centers.nodes[k]
    dT=-n["upperAirTemperature"]+n["lowerAirTemperature"]
    print(dT)
    if(dT>0):
        planet.lowerAtmosphere.sources[i]-=0.1*dT*dt/100
        planet.lowerAtmosphere.sources+=0.1*dT*dt/100/planet.lowerAtmosphere.numberOfSource
for k,T in planet.centers.nodes.data("upperAirTemperature"):
    planet.centers.nodes[k]["upperAirTemperature"]*=np.exp(-dt/1e5)
```
