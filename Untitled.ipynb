```python
@dataclass
class PlanetParameters:
    radius: float
    starTemperature: float
    starRadius: float
    starDistance: float
    

class Planet:
    def __init__(self,parameters,seed=0):
        self.seed=seed
        self.parameters=parameters
    def generateTiling(self,numberTiles,iterateRelaxation=30):
        self.centers,self.corners=Tiling(self.seed).createGraph(numberTiles,iterate=iterateRelaxation)
    def generatePlates(self,numberPlates):
        self.p=Plates(self.centers)
        self.centers,self.plates=self.p.generatePlates(numberPlates)   
        self.centers,self.plates=self.p.randomSpeed(self.plates,self.centers) 
        self.centers,self.plates=self.p.platesElevation(self.plates,self.centers,flood=0.6)
        for plate in tqdm(self.plates,desc="Computing distance matrix"):
            for (source,targets) in nx.algorithms.shortest_paths.generic.shortest_path_length(plate,weight="length"):
                plate.nodes[source]["distances"]=targets
    def computeHeight(self):
        self.centers,self.plates=self.p.computePressure(self.plates,self.centers,self.corners,self.parameters)
        #self.centers,self.plates=self.p.computeElevation(self.plates,self.centers,self.corners,self.parameters)
    @property
    def numberPlates(self):
        return len(self.plates)
    @property
    def numberTiles(self):
        return self.centers.number_of_nodes()
    def plotTiles(self,ax,contour=True,minLight=0.1):
        
        cmap_terrain = plt.cm.get_cmap('gist_earth')
        cmap_ocean = plt.cm.get_cmap('ocean')
        for i in tqdm(range(planet.centers.number_of_nodes()),desc="Plotting tiles"):    
            l=self.centers.nodes[i]["corners"]
            vertices=np.append(l,l[0])
            points=[]
            for j in vertices:
                points.append(self.corners.nodes[j]["position"])
            lat,long=Projection().toLatLong(np.array(points),"deg")
            if(self.centers.nodes[i]["elevation"]>0):
                y=np.interp(self.centers.nodes[i]["elevation"],[0,10],[0.5,1])
                color=cmap_terrain(y)
            else:
                y=np.interp(self.centers.nodes[i]["elevation"],[-10,0],[0.25,0.8])
                color=cmap_ocean(y)
            (r,g,b,a)=color
            if(self.centers.nodes[i]["temperature"]<273.15):
                r=0.9
                g=0.9
                b=1
            illumination=min(1,self.centers.nodes[i]["illumination"]+minLight)
            color=(r*illumination,g*illumination,b*illumination,1)
            ax.fill(long,lat,color=color,transform=ccrs.Geodetic()) 
            if(contour):
                ax.plot(long,lat,color="black",transform=ccrs.Geodetic(),lw=1)
    def plotPlatesSpeed(self,ax):
        centers=np.array([c for _,c in self.centers.nodes.data("center")])
        lat,long=Projection().toLatLong(centers,"deg")
        centerSpeeds=np.array([c for _,c in self.centers.nodes.data("speed")])
        slat,slong=Projection().toLatLong(centers+centerSpeeds/1000,"deg")
        for i in tqdm(range(len(long)),desc="Plotting plate velocities"):
            ax.plot([long[i],slong[i]],[lat[i],slat[i]],color="yellow",transform=ccrs.Geodetic(),lw=1)    
        ax.scatter(long,lat,color="orange",transform=ccrs.PlateCarree(),s=5,zorder=10)
    def plotPlates(self,ax):
        for (i,j) in tqdm(self.corners.edges,desc="Plotting plate borders"):
            (n1,n2)=self.corners.edges[i,j]["separates"]
            if self.centers.nodes[n1]["plate"]!=self.centers.nodes[n2]["plate"]:
                points=np.array([self.corners.nodes[i]["position"],self.corners.nodes[j]["position"]])
                lat,long=Projection().toLatLong(np.array(points),"deg")
                ax.plot(long,lat,color="red",transform=ccrs.Geodetic(),lw=2)
    def plotPressure(self,ax):
        for i,p in tqdm(self.centers.nodes.data("pressure"),desc="Plotting pressure"):
            if p is not None:
                lat,long=Projection().toLatLong(np.array([self.centers.nodes[i]["center"]]),"deg")
                if(p>0):
                    ax.scatter(long,lat,s=p/5,color="red",zorder=15,transform=ccrs.PlateCarree())
                else:
                    ax.scatter(long,lat,s=-p/5,color="blue",zorder=15,transform=ccrs.PlateCarree())
    def plotElevation(self,ax):
        points=np.array([p for _,p in self.centers.nodes.data("center")])
        elevation=np.array([e for _,e in self.centers.nodes.data("elevations")])
        lat,long=Projection().toLatLong(points,"deg")
        elevation_interp=interp2d(long,lat,elevation)
        long2d,lat2d=np.meshgrid(np.linspace(-180,180),np.linspace(-90,90))
        ax.contour(long2d,lat2d,elevation_interpolate(long2d,lat2d))
```

```python
class Illumination():
    def __init__(self,day=1,year=365,inclination=0):
        self.dayFreq=2*np.pi/day
        self.yearFreq=2*np.pi/year
        self.inclination=inclination
    def light(self,lon,lat,t):
        lat2d,t2d=np.meshgrid(lat,t,indexing="ij")
        lon2d,t2d=np.meshgrid(lon,t,indexing="ij")
        normal=np.array([np.cos(lat2d)*np.cos(lon2d),np.cos(lat2d)*np.sin(lon2d),np.sin(lat2d)])
        u=np.array([-np.cos(self.yearFreq*t2d)*np.cos(self.dayFreq*t2d)-np.sin(self.yearFreq*t2d)*np.sin(self.dayFreq*t2d)*np.cos(self.inclination),
            +np.cos(self.yearFreq*t2d)*np.sin(self.dayFreq*t2d)-np.sin(self.yearFreq*t2d)*np.cos(self.dayFreq*t2d)*np.cos(self.inclination),
            -np.sin(self.yearFreq*t2d)*np.sin(self.inclination)])
        lightMap=-(u[0]*normal[0]+u[1]*normal[1]+u[2]*normal[2])
        lightMap=np.maximum(0,lightMap)
        return lightMap
lat,long=Projection().toLatLong(np.array([p for _,p in planet.centers.nodes.data("center")]),"deg")

time=np.arange(0,365,0.1)#np.array([0,1e-3])+365/4*3+.1
light=Illumination(year=365,inclination=24/180*np.pi).light(long/180*np.pi,lat/180*np.pi,time)
light=np.trapz(light,time,axis=-1)/(time[-1]-time[0])
for i,l in enumerate(light):
    planet.centers.nodes[i]["illumination"]=l
for k,e in planet.centers.nodes.data("elevation"):
    if e<0:
        planet.centers.nodes[k]["albedo"]=0.15
    else:
        planet.centers.nodes[k]["albedo"]=0.2
for k,ill in planet.centers.nodes.data("illumination"):
    Ts=(ill*(1-planet.centers.nodes[k]["albedo"]))**(1/4)*planet.parameters.starTemperature*np.sqrt(planet.parameters.starRadius*696340e3/planet.parameters.starDistance/149597870e3)
    planet.centers.nodes[k]["temperature"]=Ts
```

```python
@njit
def toLatLong(r,ex,ey,ez):
    lat=np.zeros(r.shape[0])
    long=np.zeros(r.shape[0])
    for i in range(r.shape[0]):
        lat[i]=np.arccos(r[i][2]/np.sqrt(r[i][0]**2+r[i][1]**2+r[i][2]**2))
        long[i]=np.arctan2(r[i][1],r[i][0])
    return lat,long
def lengthSpherical(a,b):
    return np.arccos(np.sum(a*b,axis=-1))
def borderRoll(border):
    n=len(border)
    for i in range(n):
        yield((border[i],border[(i+1)%n]))
class Projection:
        def __init__(self):
            self.ex=np.array([1,0,0])
            self.ey=np.array([0,1,0])
            self.ez=np.array([0,0,1])
        def toLatLong(self,r,units="rad"):
            lat,long=toLatLong(r,self.ex,self.ey,self.ez)
            if(units=="rad"):
                return lat,long
            elif(units=="deg"):
                return (lat-np.pi/2)/np.pi*180,long/np.pi*180
```
