```python
import numpy as np
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
```

```python
class Vortex:
    def __init__(self,theta,phi,G):
        self.theta=theta
        self.phi=phi
        self.G=G
    def field(self,theta,phi):
        b=np.sin(theta)*np.sin(self.theta)*np.cos(phi-self.phi)+np.cos(theta)*np.cos(self.theta)
        alpha=np.sin(self.theta)*np.sin(phi-self.phi)
        beta=np.cos(theta)*np.sin(self.theta)*np.cos(phi-self.phi)-np.sin(theta)*np.cos(self.theta)
        v_theta=-1/4/np.pi*self.G*alpha/(1-b)
        v_phi=-1/4/np.pi*self.G*beta/(1-b)
        return np.array([v_theta,v_phi])
```
